# Interactive Storytelling with ChatGPT – Project 1 Guide

Use this notebook as a checklist and a source of **example prompts** to paste into ChatGPT. You do not need to run any code here; the prompts are for use in ChatGPT.

## Step 1: Define the story theme

**Example prompt for ChatGPT:**

> I want to create an interactive story with you. Let's make it a **fantasy adventure** (or: science fiction / mystery / adventure). You are the narrator; I will play a character and make choices. After each of your responses, give me 2–3 clear options for what to do next. Let's begin by you suggesting a short opening scene and then asking me to introduce my character.

## Step 2: Introduce main characters

**Example prompt:**

> My character is **[Name]**, a **[role/background]** with a **[personality trait]**. They have **[one sentence backstory]**. Describe how they appear in the opening scene and what the first challenge or opportunity is.

## Step 3 & 4: Begin the story and participant input

**Example prompt:**

> [Name] decides to **[describe action]**. What happens next? Give me 2–3 choices for how to proceed.

## Step 5 & 6: AI responses and branching paths

Keep responding with your character's actions or choices. Use lines like:

- *"I choose option 2."*
- *"[Name] tries to [action]. What happens?"*
- *"Before that, [Name] asks [NPC] about [topic]."*

Ask ChatGPT to **offer 2–3 options** at the end of each turn so the story branches.

## Step 7: Iteration and refinement

If the story goes off-track or you want to fix something:

> Let's retcon: instead of [what happened], **[corrected version]**. Continue from there and give me the next choices.

## Step 8 & 9: Conclusion and showcase

**Example prompt:**

> I'd like to bring this to an ending. [Describe how you want it to end, or ask for a twist.] Write the final scene and a one-paragraph summary of the full adventure so I can share it.

## Submission checklist

- [ ] Theme and genre defined
- [ ] Main character(s) introduced
- [ ] Story started with a clear opening
- [ ] Multiple turns with your inputs and AI responses
- [ ] At least one branching decision
- [ ] Conclusion (or open-ended wrap-up)
- [ ] Story transcript or summary saved for submission
